### Technical Task:

### Data Description:

The churn column indicates whether the customer left _within the last month_. Other columns include gender, dependents, monthly charges, etc. with information about the types of services each customer has.

The Telco customer churn data module is composed of 5 files in Excel format:

We have 7043 entries and 9 columns, of which `Customer ID` is the unique key for each row. We appear to have 0 NaN (null) values; however, we will still need to show `value_counts` outputs to check for values like 'nan', 'Nan', '0', 'Zero', 'n/a', etc.

Firstly, we need to rename our columns to match snake_case formatting. For this, we will use a regex function: because we know our data description in advance, we can add exceptions for abbreviations in every table:

```python
# snake_case conversion via regex (including exceptions):
def abbreviations(x):
    if 'ID' in x and 'Customer ID' not in x:
        return 'id'
    elif 'ID' in x:
        return 'user_id'
    elif 'TV' in x and 'CLTV' not in x:
        return 'streaming_tv'
    elif 'Avg Monthly GB Download' in x:
        return 'avg_monthly_gb_download'
    elif 'CLTV' in x:
        return 'cltv'
    return x

# Apply regex syntax within a function:
def col_rename(df):
    df = df.rename(columns=abbreviations)
    df = df.rename(columns=lambda x: re.sub(r"\s+", '_', re.sub(r'(?<!^)(?=[A-Z])', '', x)).lower())
    return df

demographics_df = col_rename(demographics_df)

display(demographics_df.columns)
```

Since the feature only contains `Yes` and `No` values, we can convert it to `boolean` value type:

```python
# Let us write a function which can be applied universally to other columns:
def y_n_replace(column):
    column.replace(to_replace = 'Yes', value = 1, inplace=True)
    column.replace(to_replace = 'No', value = 0, inplace=True)
    column = column.astype(bool)
    return column

# Let us also create a function which will convert the column and display its values:
def convert_display(column):
    column = y_n_replace(column)
    display(column.value_counts(), column.dtype)
    return column

demographics_df['under_30'] = convert_display(demographics_df['under_30'])
```

We have analysed and pre-processed every feature in this dataframe, let us now analyse the features' correlation (we will also need to analyse the correlation when we join our tables together, but this step may help eliminate redunant information and make our data lighter).

We will use the `phik` coefficient along the guidelines described [in this article](https://towardsdatascience.com/phik-k-get-familiar-with-the-latest-correlation-coefficient-9ba0032b37e7).

```python
# In terms of interval/continous data we shall use our int64 values:
interval_cols = ['age', 'number_of_dependents']
# user_id is not relevant, as it is the secondary key:
phik_overview = demographics_df.drop('user_id', axis=1).phik_matrix(interval_cols=interval_cols)
phik_overview.round(2)

plot_correlation_matrix(phik_overview.values,
                        x_labels=phik_overview.columns,
                        y_labels=phik_overview.index,
                        vmin=0, vmax=1, color_map='Greys',
                        title='Phik Correlation Matrix',
                        fontsize_factor=1.5,
                        figsize=(20, 18))
```

We have a high correlation between age and senior-citizen status, between age and being under 30, as well as between `dependents` and their number. Let us also analyse whether the correlation is statistically significant:

```python
significance_overview = demographics_df.drop('user_id', axis=1).significance_matrix(interval_cols=interval_cols)

# The output may take some time to run and generate:
plot_correlation_matrix(significance_overview.fillna(0).values,
                        x_labels=significance_overview.columns,
                        y_labels=significance_overview.index,
                        vmin=-5, vmax=5, title='Statistical Significance of Feature Correlation',
                        usetex=False, fontsize_factor=1.5, color_map='binary', figsize=(20, 18))
```

The correlation is most statistically significant between the status of having dependents and their number. The correlation between age and senior-citizen status or under-30 status is also fairly significant.

Similarly, `age` contains more information that `senior_citizen` and `under_30` — there are no disrepancies in our data, the two latter columns are also made redundant, since all their information is contained in `age`.

The introduction of [this article](https://towardsdatascience.com/feature-selection-why-how-explained-part-1-c2f638d24cdb) explains why having redundant and highly-correlating features can negatively impact ML models. Since our discussion above proves that we will not be losing any data by deleting the redundant features, we can safely do so:

### Data Preparation for Non-Boosting models

We will be using OHE in order to deal with categorical variables for non-boosting models. We will also need to scale our numerical features.

```python
# Applying OHE, taking care to avoid extraneous dummy-variables:
df_nb = pd.get_dummies(joint_df.drop('customer_status', axis=1), drop_first=True)

# nb = no boosting!
features_nb = df_nb
target_nb = joint_df['customer_status']


# Function for scaling features:
def scaling(numeric, features_train, features_valid, features_test):
    scaler = StandardScaler()
    # Note: we are careful to teach the `Scaler` only on the training dataset:
    scaler.fit(features_train[numeric])
    features_train[numeric] = scaler.transform(features_train[numeric])
    features_valid[numeric] = scaler.transform(features_valid[numeric])
    features_test[numeric] = scaler.transform(features_test[numeric])
    return features_train, features_valid, features_test


# Function for dataset preparation:
def dataset_prep(features, target, numeric):
    features_train, features_test_and_valid, target_train, target_test_and_valid = train_test_split(features, target, test_size = 0.3, random_state = 22)
    features_valid, features_test, target_valid, target_test = train_test_split(features_test_and_valid,
                                                                                        target_test_and_valid, test_size = 0.5, random_state=22)
    # Data randomisation:
    features_train, target_train = shuffle(features_train, target_train, random_state=22)
    features_valid, target_valid = shuffle(features_valid, target_valid, random_state=22)
    features_test, target_test = shuffle(features_test, target_test, random_state=22)
    # Note: we could have also added shuffle=True to train_test_split, since with random_state the results would have been equal
    # Scaling numeric features:
    features_train, features_valid, features_test = scaling(numeric, features_train,
                                                                 features_valid, features_test)
    return features_train, features_valid, features_test, target_train, target_valid, target_test


numeric = ['satisfaction_score', 'cltv', 'number_of_referrals', 'tenure_in_months',
           'avg_monthly_long_distance_charges', 'avg_monthly_gb_download', 'monthly_charge',
           'total_charges', 'total_refunds', 'total_extra_data_charges', 'total_long_distance_charges',
           'population', 'latitude', 'longitude', 'age', 'number_of_dependents']

features_train_nb, features_valid_nb, features_test_nb, target_train_nb, target_valid_nb, target_test_nb =\
    dataset_prep(features_nb, target_nb, numeric)

display(features_train_nb.shape, features_valid_nb.shape, features_test_nb.shape)
```

```python
pd.set_option('display.float_format', '{:.5f}'.format)

def modelling(model, features_train, target_train, features_valid, target_valid):
    start_time = time.time()
    model.fit(features_train, target_train)
    learning_time = time.time() - start_time
    predictions = model.predict(features_valid)
    prediction_time = time.time() - learning_time - start_time
    f1 = f1_score(target_valid, predictions)
    recall = recall_score(target_valid, predictions)
    return [f1, recall, learning_time, prediction_time]

# We are using practically default parameters in our models:
df_results = df_results.assign(LogisticRegression = modelling(LogisticRegression(random_state = 22, class_weight='balanced'), features_train_nb, target_train_nb, features_valid_nb, target_valid_nb))

df_results = df_results.assign(DecisionTreeClassifier = modelling(RandomForestClassifier(random_state = 22, class_weight='balanced'), features_train_nb, target_train_nb, features_valid_nb, target_valid_nb))

df_results = df_results.assign(RandomForestClassifier = modelling(RandomForestClassifier(random_state = 22, class_weight='balanced'), features_train_nb, target_train_nb, features_valid_nb, target_valid_nb))

df_results = df_results.assign(CatboostClassifier = modelling(CatBoostClassifier(random_state = 22, scale_pos_weight = scale_pos_weight_rounded, cat_features=cat_features, silent=True, allow_writing_files=False), features_train_cat, target_train_cat, features_valid_cat, target_valid_cat))

df_results = df_results.assign(LGBMClassifier = modelling(lgb.LGBMClassifier(random_state = 22, boosting_type='gbdt', is_unbalance=True, categorical_feature=cat_features), features_train_lgbm, target_train_lgbm, features_valid_lgbm, target_valid_lgbm))

display(df_results)
```

### Test-dataset Performance and Dummy Comparison

In measuring our model's performance on the testing dataset, we will compare it with a dummy classifier from the `sklearn` library in order to determine whether our model passes a basic sanity check.

For our dummy classifier, we will be using the `uniform` strategy, which means that our model will predict the `True` case at random.

This is an effective way of analysing F1–score performance, because a `most_frequent` strategy will mean that both F1 and recall will be 0 (as no true positives will be predicted). With a `constant` strategy, our recall score will always be 1, because recall = true positives / (true positives + false negatives). Another possible strategy is `stratified`, which means that we will get approximately 30% of true predictions, since the strategy respects class distribution. However, we want to simulate a situation where an algorith has no data insights for making predictions in order to justify the use of our data analysis and modelling. Therefore, we will use the `uniform` strategy.
